```
def my_func(x):
  if x:
    return x*x
  else:
    return 0
```

```
print(my_func())
```

```
def my_func(x=None):
  if x:
    return x*x
  else:
    return 0
```

```
print(my_func())
print(my_func(5))
```

```
vowels='aeiou'
sentence='I am abal'

filter_letters=[]

for letter in sentence:
  if letter not in vowels:
    filter_letters.append(letter)

print(''.join(filter_letters))
```

```
filter_letters=[letter for letter in sentence if letter not in vowels]
```
